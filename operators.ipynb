```python
#An array contains numbers from 0 to n, with one number missing.
#Find the missing number using XOR operator.


nums = [3, 0, 1,]
xor = 0
for i in range(len(nums) + 1):
    xor ^= i
for n in nums:
    xor ^= n
print(xor)




```

```python
n1 = [1,2,3,5]
n = len(n1) + 1
expected = n * (n + 1) // 2
missing = expected - sum(n1)
print(missing)
```

```python
#reverse of unsigned 32 bit
n = 5
binary = format(n,'032b')
reverse = binary[::-1]
print(int(reverse))
```

```python
#gray code
i = 5
gray = i ^ (i >> 1)
print(gray)
```

```python
#Toggle means change to the opposite state 0-1,1-0
n = 5
mask = 10
toggle_all = n ^ mask
print(toggle_all)
toggle_bit = n ^ (1 << 2)
print(toggle_bit)
toggle_range = n ^ ((1 << 2)-1)
print(toggle_range)
```
